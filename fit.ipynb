```python
# https://micropore.wordpress.com/2017/02/04/python-fit-with-error-on-y-axis/

--> DO NOT USE THIS CELL !!!

import numpy as np
from pylab import *
from scipy.optimize import curve_fit

def func(x, a, b, c):
    return a * x *x + b*x + c

# test data and error
x = np.linspace(-10, 10, 100)
y0 = - 0.07 * x * x + 0.5 * x + 2.
noise = np.random.normal(0.0, 1.0, len(x))
y = y0 + noise

# curve fit [with only y-error]
popt, pcov = curve_fit(func, x, y, sigma=1./(noise*noise))
perr = np.sqrt(np.diag(pcov))

#print fit parameters and 1-sigma estimates
print("fit parameter 1-sigma error")
print("———————————–")
for i in range(len(popt)):
    print(str(popt[i])+" +- "+str(perr[i]))

# prepare confidence level curves
nstd = 5. # to draw 5-sigma intervals
popt_up = popt + nstd * perr
popt_dw = popt - nstd * perr

fit = func(x, *popt)
fit_up = func(x, *popt_up)
fit_dw = func(x, *popt_dw)

#plot
fig, ax = plt.subplots(1)
rcParams["xtick.labelsize"] = 18
rcParams["ytick.labelsize"] = 18
rcParams["font.size"] = 20
errorbar(x, y0, yerr=noise, xerr=0, hold=True, ecolor="k", fmt="none", label="data")

xlabel("x", fontsize=18)
ylabel("y", fontsize=18)
title("fit with only Y-error", fontsize=18)
plot(x, fit, "r", lw=2, label="best fit curve")
plot(x, y0, "k-", lw=2, label="True curve")
ax.fill_between(x, fit_up, fit_dw, alpha=.25, label="5-sigma interval")
legend(loc="lower right",fontsize=18)
```

```python
#
#   Fit Gauss
#   No errors, no initialization
#

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def Gauss(x, A, mean, sigma):
    return A * np.exp(-(x - mean)**2 / 2 / sigma**2)

np.random.seed(1)

mu, sigma = 0, 1
rnormal = np.random.normal(mu, sigma, 1000)

fig, ax = plt.subplots()

bins = np.arange(-5*sigma, 5*sigma, 0.2)        # set step 0.2
# bins = np.linspace(-5*sigma, 5*sigma, 100+1)  # set the number of points (100)

y, x, _ = ax.hist(rnormal, bins=40, histtype="step")
x = 0.5 * (x[1:] + x[:-1])

popt, pcov = curve_fit(Gauss, x, y)
popt[2] = abs(popt[2])  # positive sigma
print(f"popt: {popt}")

ax.plot(x, Gauss(x, *popt))
ax.grid()

text = f"A = {popt[0]:.0f}\n$\mu$ = {popt[1]:.2f}\n$\sigma$ = {popt[2]:.2f}"
ax.text(0.05, 0.95, text, transform=ax.transAxes, fontsize=14, va='top')
```

```python
#
#   Fit Gauss
#   No errors, no initialization
#

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def fGauss(x, A, mean, sigma):
    return A * np.exp(-(x - mean)**2 / 2 / sigma**2)

np.random.seed(1)

mu, sigma = 0, 1
rnormal = np.random.normal(mu, sigma, 1000)

plt.figure()

bins = np.arange(-5*sigma, 5*sigma, 0.2)        # set step 0.2
# bins = np.linspace(-5*sigma, 5*sigma, 100+1)  # set the number of points (100)

yfit, xfit, _ = plt.hist(rnormal, bins=40, histtype="step")
xfit = 0.5 * (xfit[1:] + xfit[:-1])

popt, pcov = curve_fit(fGauss, xfit, yfit)
popt[2] = abs(popt[2])  # positive sigma
print(f"popt: {popt}")

chi2ndf = np.sum((yfit - fGauss(xfit, *popt))**2) / (len(xfit) - 3)
err = np.sqrt(yfit)
err[err == 0] = 1
chi2ndf = np.sum(((yfit - fGauss(xfit, *popt)) / err)**2) / (len(xfit) - 3)
print(f'chi2ndf: {chi2ndf}')
plt.plot(xfit, fGauss(xfit, *popt))
plt.grid()

text = f"A = {popt[0]:.0f}\n$\mu$ = {popt[1]:.2g}\n$\sigma$ = {popt[2]:.2g}\n$\chi^2$ = {chi2ndf:.2f}"
plt.text(0.05, 0.95, text, transform=plt.gca().transAxes, fontsize=14, va='top')
```
